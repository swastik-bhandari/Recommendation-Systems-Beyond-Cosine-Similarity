```python
from sklearn.metrics.pairwise import cosine_similarity 
from sklearn.metrics.pairwise import euclidean_distances
```

```python
import numpy as np
import pandas as pd

def pairwise_scaled_sim_array(pt, alpha=0.5, norm_type=2):
    """
    pt: pandas DataFrame (pivot table)
    alpha: weight between cosine similarity and norm-based term
    norm_type: 1 for L1 norm, 2 for L2 norm
    Returns: NumPy array (n_items x n_items) similar to pairwise.cosine_similarity
    """
    # Convert pivot table to NumPy
    X = pt.values.astype(float)
    
    # Compute norms
    norms = np.linalg.norm(X, ord=norm_type, axis=1)
    
    # Cosine similarity part (vectorized)
    cos_sim = X @ X.T / (norms[:, None] * norms[None, :] + 1e-10)  # avoid div by zero
    
    # Norm-based part (sum of norms)
    norm_sum = norms[:, None] + norms[None, :]
    
    # Normalize norm_sum to 01 range
    norm_sum_normalized = norm_sum / np.max(norm_sum)
    
    # Scaled similarity
    scaled_sim = alpha * cos_sim + (1 - alpha) * norm_sum_normalized
    
    # Optionally, normalize diagonal to 1 (like cosine_similarity)
    np.fill_diagonal(scaled_sim, 1.0)
    
    return scaled_sim

similarity_score_cosine = cosine_similarity(pt)
similarity_score_scaled_sim = pairwise_scaled_sim_array(pt);
```

```python
print(list(enumerate(similarity_score_scaled_sim[0])))
```

```python
def find_idx(book_name):
    for i, name in enumerate(pt.index):
        if name == book_name:
            return i
    return -1
```

```python
def recommend(book_name , similarity_score):
    index = find_idx(book_name)
    similar_items = list(enumerate(similarity_score[index]))
    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)
    similar_items = similar_items[1:7]
    similar_items = [i for i, v in similar_items]
    return similar_items
```

```python
recommended_arr_bycosine=recommend('The Da Vinci Code' , similarity_score_cosine)
recommended_arr_byscaled_sim =recommend('The Da Vinci Code' , similarity_score_scaled_sim)
```
